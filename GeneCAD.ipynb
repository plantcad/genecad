# GeneCAD: Foundation Model Genome Annotation

This notebook provides a quick start guide for running GeneCAD in Google Colab.
GeneCAD is an end-to-end genome annotation pipeline for plants and animals, powered by the DNA foundation model PlantCAD2.

For more details, visit the [GeneCAD GitHub Repository](https://github.com/plantcad/genecad).

## 1. Setup Environment
We use `uv` for fast installation and dependency management.

```python
%%bash
# Install 'uv'
curl -LsSf https://astral.sh/uv/install.sh | sh

# Download the GeneCAD repository
git clone -b refine-CrossEntropyLoss https://github.com/plantcad/genecad.git
```

## 2. Install GeneCAD
We create a virtual environment with Python 3.12 and install GeneCAD along with its dependencies.

> **Note**: This step compiles `mamba-ssm` and `causal-conv1d` which can take 5-10 minutes on Colab.

```python
%%bash
export PATH="$HOME/.local/bin:$PATH"
cd genecad

# Create a virtual environment
uv venv --python 3.12
source .venv/bin/activate

# Install GeneCAD. We set PyTorch to cu124 to match Colab's default drivers.
export PYTORCH_CUDA_INDEX="https://download.pytorch.org/whl/cu124"
bash scripts/install_release.sh
```

## 3. Run Inference
Let's run a test example on *Arabidopsis thaliana*. The TAIR12 sequence is downloaded automatically.
This will run the entire pipeline and output a GFF3 annotation file.

```python
%%bash
export PATH="$HOME/.local/bin:$PATH"
cd genecad
source .venv/bin/activate

# Run the pipeline. We specify --gpus all to use the T4/A100 GPU allocated by Colab.
genecad predict --gpus all
```

## 4. Inspect Results
GeneCAD outputs the raw predictions and a final, protein-refined GFF3 file.

```python
%%bash
ls -lh genecad/genecad_result/Athaliana_predictions/
```

```python
!head -n 20 genecad/genecad_result/Athaliana_predictions/Athaliana_GeneCAD_final.gff
```

## Optional: Run the Web UI
GeneCAD comes with a No-Code Web UI built using Gradio. You can start it here and click the public `xxxx.gradio.live` link in the output to access it in a new tab!

```python
%%bash
export PATH="$HOME/.local/bin:$PATH"
cd genecad
source .venv/bin/activate

# This will block and run the server. Click the public link to access the UI.
# Stop the cell when you are done.
genecad ui --share
```
